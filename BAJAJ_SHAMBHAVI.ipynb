```python
#QUES.9 Percentage Distribution
import pandas as pd
import json

# Load the JSON file
with open('/content/drive/MyDrive/DataEngineeringQ2.json', 'r') as file:
    data = json.load(file)

# Normalize the JSON data to flatten nested structures
df = pd.json_normalize(data, sep='_')

# Extract medicine data
medicine_data = pd.json_normalize(data,
                                  record_path=['consultationData', 'medicines'],
                                  meta=['_id', 'appointmentId'],
                                  sep='_')

# Count the number of active and inactive medicines
medicine_status_counts = medicine_data['isActive'].value_counts()

# Calculate the total number of medicines
total_medicines = medicine_status_counts.sum()

# Calculate the percentage distribution of active and inactive medicines
percentage_active = (medicine_status_counts[True] / total_medicines) * 100
percentage_inactive = (medicine_status_counts[False] / total_medicines) * 100

percentage_distribution = {
    'percentage_active': percentage_active,
    'percentage_inactive': percentage_inactive
}

percentage_distribution
```

```python
#QUES.10 Return the number of valid phone numbers found in the dataset.
import pandas as pd
import json

# Load the JSON file
with open('/content/drive/MyDrive/DataEngineeringQ2.json', 'r') as file:
    data = json.load(file)

# Normalize the JSON data to flatten nested structures
df = pd.json_normalize(data, sep='_')

# Function to check if a phone number is valid
def is_valid_mobile(phone):
    if isinstance(phone, str):
        # Remove any whitespace
        phone = phone.strip()

        # Check for +91 prefix
        if phone.startswith('+91'):
            phone = phone[3:]
        # Check for 91 prefix
        elif phone.startswith('91'):
            phone = phone[2:]

        # Check if the remaining number is within the valid range
        if phone.isdigit() and len(phone) == 10:
            number = int(phone)
            if 6000000000 <= number <= 9999999999:
                return True
    return False

# Apply the function to create the isValidMobile column
df['isValidMobile'] = df['phoneNumber'].apply(is_valid_mobile)

# Count the number of valid phone numbers
valid_phone_count = df['isValidMobile'].sum()

valid_phone_count
```

```python
import json
import numpy as np
from datetime import datetime

# Load the data
with open('/content/drive/MyDrive/DataEngineeringQ2.json') as f:
    data = json.load(f)

# Extract relevant information: birth dates and number of medicines prescribed
ages = []
num_medicines = []

current_year = datetime.now().year

for entry in data:
    patient_details = entry.get('patientDetails', {})
    birth_date = patient_details.get('birthDate')

    if birth_date:
        birth_year = datetime.fromisoformat(birth_date).year
        age = current_year - birth_year
        ages.append(age)
    else:
        ages.append(None)

    consultation_data = entry.get('consultationData', {})
    medicines = consultation_data.get('medicines', [])
    num_medicines.append(len(medicines))

# Filter out entries where age is not available
ages_filtered = [age for age in ages if age is not None]
num_medicines_filtered = [num_medicines[i] for i in range(len(num_medicines)) if ages[i] is not None]

# Calculate the Pearson correlation coefficient
pearson_correlation = np.corrcoef(ages_filtered, num_medicines_filtered)[0, 1]
print(pearson_correlation)
```
